# Does the performance in the exams vary according to gender, origin and education level of parents? 
ANSWER-----> according to the finding, it is seen that there are more number of hardworking and better female students than that of their counterparts, male. Within the group of large number female good students, it was noticed that the race, group Cs are better in reading and writing scores; whereas the D group are in math. When the correlation between the variable, parental education level, with the good student scores, it was found that those students whose parents either have associate degree or some college degree are more involved in their children's academic performance as compared to the remaining parents who have high school, masters, and bachelors. 

```python
# Calculated the Average Scores using to reach to the better performance students (by adding a number to each variables ): 
average_reading_score = dataoriginal["reading score"].mean()+1
print("AVG reading passing score:" ,average_reading_score)

average_writing_score = dataoriginal["writing score"].mean()+2
print("AVG Writing Passing score: " , average_writing_score)

average_math_score = dataoriginal["math score"].mean()+4
print("AVG Math Passing score: ", average_math_score)
```

```python
passing_writing = dataoriginal[dataoriginal["writing score"] >= average_writing_score]
passing_math = dataoriginal[dataoriginal["math score"] >= average_math_score]
passing_reading = dataoriginal[dataoriginal["reading score"] >= average_reading_score]
print("writing rows and columns:", passing_writing.shape)
print("math rows and columns:",passing_math.shape)
print("reading rows and columns:", passing_reading.shape)
```

```python
pass_math_genders = passing_math.groupby(["gender"]).count()["math score"]
print("passing math scores for each gender:", pass_math_genders)
pass_writing_genders = passing_writing.groupby(["gender"]).count()["writing score"]
print("passing writing scores fror each gender:", pass_writing_genders)
pass_reading_genders = passing_writing.groupby(["gender"]).count()["reading score"]
print("passing reading scores fror each gender:", pass_reading_genders)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=passing_math, x='parental level of education', hue='parental level of education')
plt.title('Relationship between Math Scores and Parental Level of Education')

plt.show()
plt.figure(figsize=(12, 6))
sns.histplot(data=passing_writing, x='parental level of education', hue='parental level of education')
plt.title('Relationship between Writing Scores and Parental Level of Education')

plt.show()
plt.figure(figsize=(12, 6))
sns.histplot(data=passing_reading, x='parental level of education', hue='parental level of education')
plt.title('Relationship between Reading Scores and Parental Level of Education')

plt.show()
```

```python
pass_math_Pared = passing_math.groupby(["parental level of education"]).count()["math score"]
print("passing math scores by parental level of education:", pass_math_Pared)
pass_writing_Pared = passing_writing.groupby(["parental level of education"]).count()["writing score"]
print("passing writing scores by parental level of education:", pass_writing_Pared)
pass_reading_Pared = passing_writing.groupby(["parental level of education"]).count()["reading score"]
print("passing reading scores by parental level of education:", pass_reading_Pared)
```

```python
pass_writing_Pared1 =pd.DataFrame(pass_writing_Pared)
pass_writing_Pared1.reset_index(inplace=True)
pass_reading_Pared1 =pd.DataFrame(pass_reading_Pared)
pass_reading_Pared1.reset_index(inplace=True)
pass_math_Pared1 =pd.DataFrame(pass_math_Pared)
pass_math_Pared1.reset_index(inplace=True)


x_labels = pass_math_Pared1['parental level of education']
y_values = pass_math_Pared1['math score']
plt.figure(figsize=(10, 10))
plt.bar(x_labels, y_values, align='center', width=0.2, color=['pink','red','orange','green','blue','Yellow'])
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.title('Relationship between Math Scores and Parental Level of Education')
plt.xticks(rotation=90)
plt.grid()
# Add labels to the bars
for i in range(len(x_labels)):
    plt.text(x_labels[i], y_values[i], f'{y_values[i]:.2f}')

plt.show()


x_labels = pass_reading_Pared1['parental level of education']
y_values = pass_reading_Pared1['reading score']
plt.figure(figsize=(10, 10))
plt.bar(x_labels, y_values, align='center', width=0.2, color=['pink','red','orange','green','blue','Yellow'])
plt.xlabel('Parental Level of Education')
plt.ylabel('reading Score')
plt.title('Relationship between reading Scores and Parental Level of Education')
plt.xticks(rotation=90)
plt.grid()
# Add labels to the bars
for i in range(len(x_labels)):
    plt.text(x_labels[i], y_values[i], f'{y_values[i]:.2f}')
plt.show()


x_labels = pass_writing_Pared1['parental level of education']
y_values = pass_writing_Pared1['writing score']
plt.figure(figsize=(10, 10))

plt.bar(x_labels, y_values, align='center', width=0.2, color=['pink','red','orange','green','blue','Yellow'])
plt.xlabel('Parental Level of Education')
plt.ylabel('writing Score')
plt.title('Relationship between writing Scores and Parental Level of Education')
plt.xticks(rotation=90)
plt.grid()
# Add labels to the bars
for i in range(len(x_labels)):
    plt.text(x_labels[i], y_values[i], f'{y_values[i]:.2f}')

plt.show()
```

# Does lunch type have a relationship with exam performances? How can you explain this, if any?

ANSWER------> As per the statitics show, it is the standard lunch type that is more affective for the performance of the students in the subjects (math, writing and reading). 

```python
dataoriginal.lunch.unique()
```

```python
pass_math_lunch = passing_math.groupby(["lunch"]).count()["math score"]
print("passing math scores by lunch type:", pass_math_lunch)
pass_writing_lunch = passing_writing.groupby(["lunch"]).count()["writing score"]
print("passing writing scores by lunch type:", pass_writing_lunch)
pass_reading_lunch = passing_writing.groupby(["lunch"]).count()["reading score"]
print("passing reading scores by lunch type:", pass_reading_lunch)

pass_writing_lunch1 =pd.DataFrame(pass_writing_lunch)
pass_writing_lunch1.reset_index(inplace=True)
pass_reading_lunch1 =pd.DataFrame(pass_reading_lunch)
pass_reading_lunch1.reset_index(inplace=True)
pass_math_lunch1 =pd.DataFrame(pass_math_lunch)
pass_math_lunch1.reset_index(inplace=True)

x_labels = pass_math_lunch1['lunch']
y_values = pass_math_lunch1['math score']

plt.bar(x_labels, y_values, align='center', width=0.2)
plt.xlabel('lunch')
plt.ylabel('Math Score')
plt.title('Relationship between Math Scores and lunch')
plt.xticks(rotation=90)
plt.grid()

plt.show()


x_labels = pass_reading_lunch1['lunch']
y_values = pass_reading_lunch1['reading score']

plt.bar(x_labels, y_values, align='center', width=0.2)
plt.xlabel('lunch')
plt.ylabel('reading Score')
plt.title('Relationship between reading Scores and lunch')
plt.xticks(rotation=90)
plt.grid()

plt.show()


x_labels = pass_writing_lunch1['lunch']
y_values = pass_writing_lunch1['writing score']

plt.bar(x_labels, y_values, align='center', width=0.2)
plt.xlabel('lunch')
plt.ylabel('writing Score')
plt.title('Relationship between writing Scores and lunch')
plt.xticks(rotation=90)
plt.grid()

plt.show()
```
