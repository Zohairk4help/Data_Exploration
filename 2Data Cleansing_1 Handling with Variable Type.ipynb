```python
def discretization(Rating):
    if Rating <= 2:
        return("1-2")
    elif (Rating > 2) and (Rating < 4):
        return("2-4")
    elif (Rating > 4) and (Rating < 6):
        return("4-6")
    elif (Rating > 6) and (Rating < 8):
        return("6-8")
    else:
        return("x>8")
data['Rating_range'] = data['Rating'].apply(discretization)
data
```

```python
##### There are 13 columns as shown above in cell 11 and 1000 rows
```

```python
data.groupby("Rating_range")["Rating"].count().sort_values(ascending=False)
```

```python
def discretization(Year):
    if Year <= 2006:
        return("X < 2006")
    elif (Year >= 2006) and (Year <= 2008):
        return("2006-2008")
    elif (Year >= 2008) and (Year <= 2010):
        return("2008-2010")
    elif (Year >= 2010) and (Year <= 2012):
        return("2010-2012")
    elif (Year >= 2012) and (Year <= 2014):
        return("2012-2014")
    else:
        return("x > 2014")
data['Year_Range'] = data['Year'].apply(discretization)
data
```

```python
##### There are 14 columns as shown above in cell 15
```

```python
data.groupby("Year_Range")["Year"].count().sort_values(ascending=False)
```

```python
data.groupby("Year_Range")["Year"].count().sort_values(ascending=False)
```

###### By the example above, it is clearly seen that the column 'Year' from the dataframe, Data, becomes categorical even though it was continuos/numeric before.  

###### It is true that while continuous data is generally preferred over discrete data, discrete is the preferred data type although continuous data is available for the same characteristic. It is because as it is seen in the example above for the column "year", we can easily divide the highest ratings of the movies in the exact year.
