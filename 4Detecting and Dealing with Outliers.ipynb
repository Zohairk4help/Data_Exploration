```python
data2= dataoriginal
data2["TOTAL_REVENUE_imputed"]=data2["TOTAL_REVENUE"].fillna(data2["TOTAL_REVENUE"].mean())
data2["TOTAL_EXPENDITURE_imputed"]=data2["TOTAL_EXPENDITURE"].fillna(data2["TOTAL_EXPENDITURE"].mean())
data2.isnull().sum()
```

```python
data2= dataoriginal
data2["TOTAL_REVENUE_imputed"]=data2.TOTAL_REVENUE_imputed.interpolate(method='polynomial', order=2, limit_direction ='forward')
data2["TOTAL_EXPENDITURE_imputed"]=data2.TOTAL_EXPENDITURE_imputed.interpolate(method='polynomial', order=2, limit_direction ='forward')

data2
```

```python
from scipy.stats import zscore
zscores = zscore(data2["TOTAL_EXPENDITURE_imputed"])
for i in range(1,5):
    print("The score threshold is: {}".format(i))
    print("The indices of the outliers of TOTAL_EXPENDITURE_imputed:")
    print(np.where(zscores > i))
    print("Number of outliers is: {}".format(len((np.where(zscores > i)[0]))))
    print('--'*40)
    
print("x-x-"*29)

from scipy.stats import zscore
zscores = zscore(data2["TOTAL_REVENUE_imputed"])
for i in range(1,5):
    print("The score threshold is: {}".format(i))
    print("The indices of the outliers of TOTAL_REVENUE_imputed:")
    print(np.where(zscores > i))
    print("Number of outliers is: {}".format(len((np.where(zscores > i)[0]))))
    print('--'*40)
```

```python
# Turkey Method to find outliers
q75, q25 = np.percentile(data2['TOTAL_EXPENDITURE'].dropna(),[75 ,25])
IQR = q75 - q25
threshold_values=[]
for threshold_value in np.arange(1,10,0.5):
    min_value = q25 - (IQR*threshold_value)
    max_value = q75 + (IQR*threshold_value)
    outlier_value_number = len((np.where((data2['TOTAL_EXPENDITURE'].dropna() > max_value)| (data2['TOTAL_EXPENDITURE'].dropna() < min_value))[0]))
    threshold_values.append({'threshold_value': threshold_value,'outlier_value_number': outlier_value_number})
pd.DataFrame(threshold_values)
```

```python
q75, q25 = np.percentile(data2['TOTAL_REVENUE'].dropna(),[75 ,25])
IQR = q75 - q25
threshold_values=[]
for threshold_value in np.arange(1,10,0.5):
    min_value = q25 - (IQR*threshold_value)
    max_value = q75 + (IQR*threshold_value)
    outlier_value_number = len((np.where((data2['TOTAL_REVENUE'].dropna() > max_value)| (data2['TOTAL_REVENUE'].dropna() < min_value))[0]))
    threshold_values.append({'threshold_value': threshold_value,'outlier_value_number': outlier_value_number})
pd.DataFrame(threshold_values)
```

```python
#Removing outliers using the threshold value found above in cells:
# Set a threshold value for the high end of the price range
threshold = data2['TOTAL_REVENUE'].quantile(0.075)

# Remove any houses that are above the threshold
df = data2[data2['TOTAL_REVENUE'] <= threshold]

# Display the updated dataset
df
```

# Method: Transforming the Outliers or triming the outliers

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.hist(data2['TOTAL_REVENUE'].dropna())
plt.title("TOTAL_REVENUE before transformation")
plt.subplot(1,2,2)
plt.hist(np.log(data2['TOTAL_REVENUE'].dropna()))
plt.title("TOTAL_REVENUE after Log Transformation")
plt.show()
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.hist(data2['TOTAL_EXPENDITURE'].dropna())
plt.title("TOTAL_EXPENDITURE before transformation")
plt.subplot(1,2,2)
plt.hist(np.log(data2['TOTAL_EXPENDITURE'].dropna()))
plt.title("TOTAL_EXPENDITURE after Log Transformation")
plt.show()
```

```python
q75_log, q25_log = np.percentile(np.log(data2['TOTAL_EXPENDITURE'].dropna()), [75 ,25])
IQR_log = q75_log - q25_log
log_threshold_values=[]
for threshold_value in np.arange(1,5,0.5):
    min_value_log = q25_log - (IQR_log*threshold_value) 
    max_value_log = q75_log + (IQR_log*threshold_value)
    min_value = q25 - (IQR*threshold_value) 
    max_value = q75 + (IQR*threshold_value)
    outlier_value_number_log = len((np.where((np.log(data2['TOTAL_EXPENDITURE'].dropna()) > max_value_log)| 
                                       (np.log(data2['TOTAL_EXPENDITURE'].dropna()) < min_value_log))[0]))
    outlier_value_number = len((np.where((data2['TOTAL_EXPENDITURE'].dropna() > max_value)| 
                                       (data2['TOTAL_EXPENDITURE'].dropna() < min_value))[0]))
    log_threshold_values.append({'threshold_value': threshold_value,
                                 'outlier_value_number' : outlier_value_number,
                                 'outlier_value_number_log TOTAL_EXPENDITURE': outlier_value_number_log })
pd.DataFrame(log_threshold_values)
```
