###### Noticing that the data has a time dimension (year). For this assignment, i forgot about time and treated all the observations as if they're from the same year. Chose a strategy to deal with the missing values for each variable. Finding for which variables would be filling in the missing values with some value make sense? and For which might tossing out the records entirely make sense?

The stats of the top 4 columns that has the most missing values before handling NaN in it.

```python
data2= data
data2["AVG_READING_4_SCORE_imputed"]=data2["AVG_READING_4_SCORE"].fillna(data2["AVG_READING_4_SCORE"].mean())
data2["AVG_READING_8_SCORE_imputed"]=data2["AVG_READING_8_SCORE"].fillna(data2["AVG_READING_8_SCORE"].mean())
data2["AVG_MATH_8_SCORE_imputed"]=data2["AVG_MATH_8_SCORE"].fillna(data2["AVG_MATH_8_SCORE"].mean())
data2["AVG_MATH_4_SCORE_imputed"]=data2["AVG_MATH_8_SCORE"].fillna(data2["AVG_MATH_8_SCORE"].mean())
data2
```

```python
miss_imp1=data["AVG_READING_4_SCORE_imputed"].isnull().sum()
miss_imp2=data["AVG_READING_8_SCORE_imputed"].isnull().sum()
miss_imp3=data["AVG_MATH_8_SCORE_imputed"].isnull().sum()
miss_imp4=data["AVG_MATH_4_SCORE_imputed"].isnull().sum()
print(miss_imp1)
print(miss_imp2)
print(miss_imp3)
print(miss_imp4)

print("=======================")
print("this is the sum of null values in the imputed comlumns where there were the most null values.")
```

## Took into account the time factor. Replicated the second answer, but this time, filled in the missing values by using a statistic that is calculated within the year of the observation. For example, if there is a filling in a missing value for a variable with the mean of that variable, icalculated the mean by using only the observations for that specific year.


```python
## To compare the effects of fill missing values with mean in the top 4 columns where missing values is the highest
compare_df=data2.filter(["AVG_READING_8_SCORE", "AVG_MATH_8_SCORE_imputed","AVG_READING_4_SCORE","AVG_READING_4_SCORE_imputed", "AVG_MATH_8_SCORE","AVG_MATH_8_SCORE_imputed", "AVG_MATH_4_SCORE","AVG_MATH_4_SCORE_imputed"], axis=1)
compare_df.describe()
```

After filling the misisng values with mean values in the columns, it is clearly seen that there is some difference in the standard deviation and mean of the columns: 	AVG_READING_8_SCORE, AVG_READING_8_SCORE_imputed, AVG_MATH_4_SCORE and AVG_MATH_4_SCORE_imputed 

the only columns that has no effects on their mean, but not significant values are: AVG_MATH_8_SCORE, AVG_MATH_8_SCORE_imputed, AVG_READING_8_SCORE, and AVG_READING_8_SCORE_imputed.
